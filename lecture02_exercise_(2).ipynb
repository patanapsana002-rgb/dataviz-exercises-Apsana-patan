## Setup — Run this cell first


```python
import pandas as pd
import numpy as np

# World Happiness Report 2023 — representative data
# Source: https://www.kaggle.com/datasets/ajaypalsinghlo/world-happiness-report-2023

df = pd.read_csv('/content/world_happiness_2023.csv')
df.columns = ['Country','Region','Happiness_Score','GDP','Social_Support',
              'Life_Expectancy','Freedom','Generosity','Corruption']


print(f"Dataset: {len(df)} countries, {len(df.columns)} columns")
print(df.head())
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Explore the dataset before you start
print("Regions in dataset:")
print(df['Region'].value_counts())
print("\nScore range:", df['Happiness_Score'].min(), "–", df['Happiness_Score'].max())
print("\nBottom 10 countries:")
print(df.nsmallest(10, 'Happiness_Score')[['Country','Region','Happiness_Score']])
```

## Task 1 — Regional Comparison Bar Chart

**What to build:** A horizontal bar chart showing the **average happiness score by region**, sorted from highest to lowest.

**Requirements:**
- Horizontal orientation (category names are long)
- Sorted by score, descending (so the happiest region is at the top)
- Zero baseline on x-axis
- At least one design choice that goes beyond the Plotly default (colour, annotation, labels, etc.)
- An insight title that answers: *which region stands out and why does it matter?*

**Hint:** Use `df.groupby('Region')['Happiness_Score'].mean()` to compute the averages.


```python
import plotly.express as px

# Average happiness by region
region_avg = (
    df.groupby('Region')['Happiness_Score']
    .mean()
    .sort_values()
    .reset_index()
)

# Plot
fig1 = px.bar(
    region_avg,
    x='Happiness_Score',
    y='Region',
    orientation='h',
    title='Average Happiness Score by Region',
    labels={'Happiness_Score': 'Average Happiness Score'}
)

fig1.update_layout(template='plotly_white')
fig1.show()
```

Average happiness varies significantly across regions, highlighting global inequality in well-being.
Western Europe and similar regions tend to score higher, reflecting stronger economies and social systems.
Lower scores in regions like Sub-Saharan Africa suggest structural challenges affecting quality of life.
Overall, geography plays a major role in shaping happiness outcomes.

## Task 2 — Bottom vs. Top: A Contrast Story

**What to build:** A bar chart that highlights the **gap between the happiest and least happy countries**, focusing on a specific insight.

**Requirements:**
- Show the **top 8 AND bottom 8 countries** together (16 bars total)
- Use **colour** to distinguish the two groups (not Plotly's default rainbow)
- Add a **visual separator or annotation** that emphasises the gap
- Insight title that tells the story of the gap

**Hint:** Use `pd.concat([df.nlargest(8,'Happiness_Score'), df.nsmallest(8,'Happiness_Score')])` to get both groups.

**Stretch goal:** Add a vertical reference line showing the global average.


```python
# Top 8 happiest countries
top8 = df.nlargest(8, 'Happiness_Score').copy()
top8['Group'] = 'Top 8'

# Bottom 8 least happy countries
bottom8 = df.nsmallest(8, 'Happiness_Score').copy()
bottom8['Group'] = 'Bottom 8'

# Combine
contrast_df = pd.concat([top8, bottom8])

# Sort for clean visualization
contrast_df = contrast_df.sort_values('Happiness_Score')

# Plot
fig2 = px.bar(
    contrast_df,
    x='Happiness_Score',
    y='Country',
    color='Group',
    orientation='h',
    title='Top 8 vs Bottom 8 Countries by Happiness Score',
    labels={'Happiness_Score': 'Happiness Score'}
)

fig2.update_layout(template='plotly_white')
fig2.show()

```

The gap between the top and bottom countries is substantial, showing uneven global life satisfaction.
Top countries consistently benefit from strong economies, social support, and personal freedoms.
Bottom countries often face limitations in these same areas, impacting overall happiness.
This contrast emphasizes how multiple factors combine to influence national well-being.

## Done? Stretch Goal

If you finish both tasks with time to spare, try this:

**Task 3 (stretch):** Build a **grouped bar chart** comparing 2 sub-factors (e.g. `GDP_per_capita` and `Freedom`) across the 5 most populated regions. Use colour meaningfully and write an insight title.

Regions to include: `'Western Europe'`, `'Latin America'`, `'East Asia'`, `'Sub-Saharan Africa'`, `'South Asia'`


```python

# Step 1: Filter required regions
regions = [
    'Western Europe',
    'Latin America',
    'East Asia',
    'Sub-Saharan Africa',
    'South Asia'
]

filtered_df = df[df['Region'].isin(regions)].copy()

# Step 2: Compute regional averages
region_means = (
    filtered_df.groupby('Region')[['GDP', 'Freedom']]
    .mean()
    .reset_index()
)

# Step 3: Reshape for grouped bar chart
long_df = region_means.melt(
    id_vars='Region',
    value_vars=['GDP', 'Freedom'],
    var_name='Factor',
    value_name='Value'
)

# Step 4: Plot
import plotly.express as px

fig3 = px.bar(
    long_df,
    x='Region',
    y='Value',
    color='Factor',
    barmode='group',
    title='Wealth vs Freedom Across Regions: Economic Strength Doesn’t Always Equal Personal Freedom'
)

fig3.update_layout(template='plotly_white')

fig3.show()
```

Regions with higher GDP generally show higher levels of freedom, indicating a positive relationship.
However, the pattern is not perfectly consistent across all regions.
Some regions achieve moderate freedom despite lower economic output.
This suggests that economic growth alone does not fully determine personal freedom.
